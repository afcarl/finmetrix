## Cox-Ingersoll-Ross model

$$
dr_{t}=\kappa\left(\mu-r_{t}\right)dt+\sigma\sqrt{r_{t}}dW_{t}
$$

Below we define the class for CIR process together with a few estimation methods applicable for it.<p>
More details of estimation of diffusion processes can be found [here]({filename}/diffusion-estimation.md).

```python
class CIR(object):
    
    def __init__(self):
        self.names = ['kappa','mu','sigma']
    
    def transform_param(self, theta):
        kappa, mu, sigma = theta
        self.K0 = kappa * mu
        self.K1 = -kappa
        self.H0 = 0.
        self.H1 = sigma ** 2
    
    def is_valid(self, theta):
        kappa, mu, sigma = theta
        return 2 * kappa * mu - sigma ** 2 > 0
    
    def estimate_ols_euler(self, x):
        import statsmodels.api as sm
        Y = (x[1:] - x[:-1]) / x[:-1] ** .5
        X1 = self.h / x[:-1] ** .5
        X2 = - self.h * x[:-1] ** .5
        X = np.vstack([X1, X2]).T
        results = sm.OLS(Y, X).fit()
        kappa = results.params[1]
        mu = results.params[0] / results.params[1]
        sigma = (results.mse_resid / self.h) ** .5
        return np.array([kappa, mu, sigma])

    def exact_likelihood(self, theta, x):
        from scipy.stats import ncx2
        kappa, mu, sigma = theta
        e = exp( - kappa * self.h )
        c = 2 * kappa / sigma ** 2 / (1 - e)
        q = 2 * kappa * mu / sigma ** 2 - 1
        v = 2 * c * x[1:]
        df = 2 * (q + 1)
        nc = 2 * c * x[:-1] * e
        l = ncx2.logpdf(v, df = df, nc = nc) + log(2 * c)
        return - l[np.isfinite(l)].mean()
    
    def estimate_mle(self, x, theta0):
        res = minimize(self.exact_likelihood, theta0, args = (x,),
                       method = 'Nelder-Mead', options = {'disp': False})
        return res.x
```

## Estimate objective parameters of the model

```python
cir = CIR()
cir.h = 1.
theta_ols = cir.estimate_ols_euler(r)
theta_mle = cir.estimate_mle(r, theta_ols)

print 'OLS = ', theta_ols
print 'MLE = ', theta_mle

print 'Feller condition: ', cir.is_valid(theta_mle)
```

## Transform analysis

More details on how transforms of stochastic processes work can be found [here]({filename}/vasicek-bond-pricing.md).

```python
import scipy.integrate as si
import functools

class Transform(object):
    
    def __init__(self, Model):
        self.Model = Model
    
    def df(self, t, x):
        x = np.atleast_2d(x)
        K0 = np.atleast_1d(self.Model.K0)
        K1 = np.atleast_2d(self.Model.K1)
        H0 = np.atleast_2d(self.Model.H0)
        H1 = np.atleast_3d(self.Model.H1)
        da = self.R0 - K0 * x[:,1] - .5 * x[:,1].T * H0 * x[:,1]
        db = self.R1 - K1.T * x[:,1] - .5 * x[:,1].T * H1 * x[:,1]
        return np.array([-da.flatten(), -db.flatten()]).flatten()
    
    def f(self):
        import itertools
        dt = np.min([1e-0, np.min(self.h)])
        r = si.complex_ode(self.df)
        sol = []
        for v, h in itertools.product(np.atleast_1d(self.u),
                                      np.atleast_1d(self.h)):
            x0 = [0., v * 1j]
            r.set_initial_value(x0, 0)
            while r.successful() and r.t < h:
                r.integrate(r.t + dt)
            sol.append(r.y)
        return np.array(sol).T

    def yields(self, x0, h):
        self.R0, self.R1 = 0., 1.
        self.u, self.h = 0., h
        A = - self.f()[0].real / h
        B = - self.f()[1].real / h
        if isinstance(x0, float):
            return A + B * x0
        else:
            return A + B * np.atleast_2d(x0).T
```

## Compute yields given the risk price

We consider two optons for market price of risk:
$$
\Lambda_{1t}=\lambda_{1}r_{t}^{-1/2},\quad\Lambda_{2t}=\lambda_{2}r_{t}^{1/2}.
$$

This produces two risk-neutral models for the short interest rate:
\begin{eqnarray*}
dr_{t} & = & \kappa\left[\left(\mu-\frac{\lambda_{1}\sigma}{\kappa}\right)-r_{t}\right]dt+\sigma r_{t}^{1/2}dW_{t}^{Q}.\\
dr_{t} & = & \left(\kappa-\lambda_{2}\sigma\right)\left[\frac{\kappa\mu}{\kappa-\lambda_{2}\sigma}-r_{t}\right]dt+\sigma r_{t}^{1/2}dW_{t}^{Q}.
\end{eqnarray*}

Below we implement two functions that change objective parameters into risk-neutral ones, and apply numerical transform method to compute model implied yields given the current short rate, maturities, and the risk price parameter.

```python
def yields1(x0, H, price):
    kappa, mu, sigma = theta_mle
    # Feller condition for CIR:
    # 2 * kappa * muQ - sigma ** 2 > 0
    # muQ > sigma ** 2 / kappa / 2
    # mu - price * sigma / kappa > sigma ** 2 / kappa / 2
    # price < (mu - sigma ** 2 / kappa / 2) * (kappa / sigma)
    max_price = (mu - sigma ** 2 / kappa / 2) * (kappa / sigma)
    if price >= max_price:
        print 'The price is too big!!! %.2f' % price, '%.2f' % max_price
    muQ = mu - price * sigma / kappa
    thetaQ = [kappa, muQ, sigma]
    cir.transform_param(thetaQ)
    tr = Transform(cir)
    return tr.yields(x0 = x0, h = H)

def yields2(x0, H, price):
    kappa, mu, sigma = theta_mle
    # Feller condition for CIR:
    # 2 * kappaQ * muQ - sigma ** 2 > 0
    # 2 * kappa * mu - sigma ** 2 > 0
    # kappaQ = kappa - price * sigma > 0
    # price < kappa / sigma
    max_price = kappa / sigma
    if price >= max_price:
        print 'The price is too big!!! %.2f' % price, '%.2f' % max_price
    kappaQ = kappa - price * sigma
    muQ = mu * kappa / kappaQ
    thetaQ = [kappa, muQ, sigma]
    cir.transform_param(thetaQ)
    tr = Transform(cir)
    return tr.yields(x0 = x0, h = H)
```

Illustrate how these functions work:

```python
price = -.5
H = np.linspace(10, 100, 1e1)
x0 = 1.5
y1 = yields1(x0, H, price)
y2 = yields2(x0, H, price)

plt.plot(H, y1.T, label = 'Model 1')
plt.plot(H, y2.T, label = 'Model 2')
plt.legend()
plt.show()
```

## Calibrate the risk price

Below we define the function that, given the risk price parameter vector, computes two version of the yields, compares them to the data, and returns Root Mean Squared Error (RMSE) indexed by the input parameter.

```python
def rmse(price):
    RMSE1, RMSE2 = [], []
    for p in np.atleast_1d(price):
        y1 = yields1(x0, H, p)
        y2 = yields2(x0, H, p)
        RMSE1.append(((df - y1) ** 2).mean().mean() ** .5)
        RMSE2.append(((df - y2) ** 2).mean().mean() ** .5)
    data = {'RMSE1' : RMSE1, 'RMSE2' : RMSE2}
    return ps.DataFrame(data, index = ps.Index(price, name = 'price'))
```
