## Heston model

Assume that the log-price evolves according to the following SV model:

\begin{eqnarray*}
d\log S_{t} & = & \left(\mu-\frac{1}{2}V_{t}\right)dt+\sqrt{V_{t}}dW_{1t},\\
dV_{t} & = & \kappa_{v}\left(\theta_{v}-V_{t}\right)dt+\sigma_{v}\sqrt{V_{t}}dW_{2t},
\end{eqnarray*}
with $Corr\left(dW_{1t},dW_{2t}\right)=\rho$.

Below we define the class for this model. First of all we need to define its drif and diffusion functions.

```python
class Heston(object):
    
    def __init__(self):
        self.eps = None
        self.names = ['mu_r','kappa','mu_v','eta','rho']
    
    def drift_S(self, x, theta):
        """Price drift."""
        mu_r, kappa, mu_v, eta, rho = theta
        # If vol is annualized:
        #return mu_r - .5 * (x/252)
        # If std is given in percentages:
        return mu_r - .5 * (x/1e4)
    
    def drift_V(self, x, theta):
        """Volatility drift."""
        mu_r, kappa, mu_v, eta, rho = theta
        return kappa * (mu_v - x)
    
    def diff_S(self, x, theta):
        """Price diffusion."""
        # If vol is annualized:
        #return (x/252) ** .5
        # If std is given in percentages:
        return (x/1e4) ** .5
    
    def diff_V(self, x, theta):
        """Volatility diffusion."""
        mu_r, kappa, mu_v, eta, rho = theta
        return eta * x ** .5
    
    def drift(self, x, theta):
        """Bivariate drift."""
        return np.vstack([self.drift_S(x[:,1], theta), self.drift_V(x[:,1], theta)]).T
    
    def diff(self, x, theta):
        """Bivariate diffusion."""
        return np.vstack([self.diff_S(x[:,1], theta), self.diff_V(x[:,1], theta)]).T
    
    def is_valid(self, theta):
        """Check Feller condition."""
        mu_r, kappa, mu_v, eta, rho = theta
        return 2 * kappa * mu_v - eta ** 2 > 0
```

## Simulation methods

Now that we have a basic description of the Heston model, we can write the method to simulate some data using Euler discretization scheme.

This function does a couple of tricks to make errors nicer and reduce Monte-Carlo variance.

```python
def nice_errors(self, e, sdim):
    """Normalize the errors and apply antithetic sampling.
    
    sdim : which dimention corresponds to simulation instances?
    """
    e -= e.mean(sdim, keepdims = True)
    e /= e.std(sdim, keepdims = True)
    e = np.concatenate((e, -e), axis = sdim)
    return e

Heston.nice_errors = nice_errors
```

Now the simulation function itself.

```python
def simulate(self, x0, theta, h, T, M, S):
    """Simulate bivariate diffusion.
    
    x0 : initial value of the process
    theta : parameters of the model
    h : time length of the unit interval
    T : total time length of the series
    M : number of intermediate steps for Euler discretization
    S : number os simulations
    """
    self.h = h
    # How many actual data points to generate?
    N = int(float(T) / h)
    # Make sure that rho is always the last parameter
    rho = theta[-1]
    # Covariance matrix for the errors
    cov = [[1., rho],[rho, 1.]]
    # Generate random normal variables only in case if they don't exist
    if self.eps is None or np.shape(self.eps) != (N, M, S, 2):
        e = np.random.multivariate_normal(np.zeros(2), cov, size = (N, M, S/2))
        self.eps = self.nice_errors(e, 2)
    
    # Initialize the simulation array
    x = np.ones((N, S, 2)) * x0
    
    def sim(z, e):
        """Iterates from previous value to the next given the shock."""
        return z + self.drift(z, theta) * (h/M) + self.diff(z, theta) * (h/M) ** .5 * e
    # Itarete for each unit time interval
    for t in xrange(N-1):
        x[t+1] = reduce(sim, self.eps[t], x[t])
    # Break down generic x into meaningful data
    self.assign_data(x)

Heston.simulate = simulate
```

The simulation method returns some generic x array. We need to extract prices, returns, and volatilities from that.

```python
def assign_data(self, x):
    """Break down generic x into meaningful data."""
    # Log-price
    self.logS = x[:,:,0]
    # Volatility
    self.V = x[:,:,1]
    # Log-returns
    self.R = np.zeros_like(self.logS)
    self.R[1:,] = self.logS[1:,] - self.logS[:-1,]

Heston.assign_data = assign_data
```

## PF methods specific to the model

To continue with the Particle Filter we need to define two methods for the model class. First will simulate only latent process. The second will compute the density of the observed data point given the simulated particles.

```python
def sim_latent(self, x0, theta, M, S):
    """Simulate latent volatility only."""
    def sim(z, e):
        loc = self.drift_V(z, theta) * (self.h / M)
        scale = self.diff_V(z, theta) * (self.h / M) ** .5
        # Fix simulation scheme!!! Taking the absolute is the temporary measure.
        return np.abs(z + loc + scale * e)
    e = np.random.normal(size = (M, S/2))
    e = self.nice_errors(e, 1)
    x = np.ones((M,S)) * x0
    for t in xrange(M-1):
        x[t+1] = sim(x[t], e[t+1])
    return x

def density_observed(self, y, x, theta):
    """Evaluate density of the observed conditional on particles."""
    loc = self.drift_S(x, theta).mean(0)
    scale = (self.diff_S(x, theta) ** 2).mean(0) ** .5
    return norm.pdf(y, loc = loc, scale = scale)

Heston.sim_latent = sim_latent
Heston.density_observed = density_observed
```

## Particle filter class

```python
class ParticleFilter(object):
    
    def __init__(self, Model, observed, start, theta):
        # Only two methods of the Model object are used
        self.Model = Model
        self.theta = theta
        self.x0 = start
        self.N = len(observed)
        self.observed = observed
    
    def resample(self, x, w):
        """Resample x from multinomial distribution with weights w."""
        draw = np.random.multinomial(len(w), w)
        return np.repeat(x, draw)
    
    def SIR(self, M, S):
        """Sampling-Importance Resampling method."""
        x = np.ones((self.N, S)) * self.x0
        for n in xrange(self.N-1):
            latent = self.Model.sim_latent(x[n], self.theta, M, S)
            p = self.Model.density_observed(self.observed[n+1], latent, self.theta)
            w = p / p.sum()
            x[n+1] = self.resample(latent[-1], w)
        return x
```

## Using the PF class

Initialize the PF instance.

```python
pf = ParticleFilter(hes, R, mu_v, theta_true)
```

Run SIR PF method to filter out the volatility.

```python
# M - number of sub-intervals inside unit interval
# S - number os simulations
M, S = 100, 200
%time V_latent = pf.SIR(M, S)
```

Plot the result.

```python
plt.figure(figsize = (12,5))
plt.plot(V_latent, color = 'black', alpha = .005)
plt.plot(V, label = 'True')
plt.plot(V_latent.mean(1), label = 'Filtered', color = 'red')
plt.legend()
plt.xlabel('Days')
plt.ylabel('Volatility')
plt.show()
```
